```python
# El script lee datos desde el archivo perceptron.csv
import pandas as pd
import math
import numpy as np

# Read Data
df = pd.read_csv("perceptron.csv")
headers_data = ['VALUE1','VALUE2']
target_name = 'TARGET'
weights = np.array([-0.7, 0.2])
bias = np.array([0.5])
```

```python
def hardlim(weights, bias, data):
    x = float(np.dot(weights,data)+bias)
    if x >= 0:
        return 1
    else:
        return 0

def update(weights, bias, data, e):
    weights_n = weights + e*data.T
    bias_n = bias + e
    return weights_n,bias_n
```

```python
change = True
iteration = 1
while change:
    print('Iteracion {}'.format(iteration))
    i = 0
    change = False
    for row in df['TARGET']:
        print('Probando con el dato {}'.format(i+1))
        data = []
        for name in headers_data:
            data.append([df[name][i]])
        data_n = np.array(data)
        target = df[target_name][i]
        res = hardlim(weights, bias, data_n)
        print('Resultado del Hardlim es {}'.format(res))
        if target != res:
            print('Se actualizan los pesos y bias')
            weights, bias = update(weights, bias, data_n, target-res)
            change = True
            print('Nuevos Pesos {}'.format(weights))
            print('Nuevo Bias {}'.format(bias))
        i += 1
    iteration += 1
    
    
print("\nResultados Finales")
print('Pesos {}'.format(weights))
print('Bias {}'.format(bias))
print('\nPuntos de Corte en los ejes')
b = float(bias[0])
i = 0
for head in headers_data:
    w = float(weights[0,i])
    print('Con {head} es {res: .4f}'.format(head=head,res= -b/w))
    i += 1
#print('Eje X: {0: .2f}'.format(- float(bias[0])/-float(weights[0]))
```
